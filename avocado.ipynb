```python
#we will use this to graph all the regions in our plot
df['region'].unique()
"""
Steps for plotting the regions:
1)currently our regions values in rows
2)we want to change it so that the column header are the regions and the values are our moving average of n and the index is the date
"""
```

```python
graph_df=pd.DataFrame()
for region in df['region'].unique()[:16]:
    print(region)
    #in df under the region column, copy all the rows of a specific region by our iteration
    region_df = df.copy()[df['region']==region]
    #change the index of the region to  "Date"
    region_df.set_index("Date",inplace=True)
    #sort the "Date" of our region
    region_df.sort_index(inplace=True)
    #create a new column with our region and assign the values of our moving average
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    #if there is nothing within our graph
    if graph_df.empty:
        #all we want is our moving average becaus we will be graphing that
        #however this won't work since this will return a series
#         graph_df=region_df['price25ma']
        #this will work, we can also specify other columns by adding the name of the column within the list with 'price25ma'
        graph_df=region_df[[f'{region}_price25ma']]
    else:
        #joining dataframes by an index
        #this will only work if 2 dataframes have the same column name
        graph_df=graph_df.join(region_df[f'{region}_price25ma'])
        
```

```python
"""
Near the end things will start to slow down, why?
This is because  the 'type' column  has 2 unique values(organic & conventional),
so when pandas in assigning and checking the indexes, there will multiple indexes that are the same
"""
print((df['type'].unique()))
df.tail()
```

```python
#This is how we fix it
df = pd.read_csv("avocado.csv")
df=df.copy()[df['type']=='organic']
df["Date"] = pd.to_datetime(df["Date"])


#sorting by a dataframe by a column rather than the index
df.sort_values(by='Date',ascending=True,inplace=True)

df.head()
```

```python
graph_df=pd.DataFrame()
for region in df['region'].unique():
    print(region)
    region_df = df.copy()[df['region']==region]
    region_df.set_index("Date",inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df=region_df[[f'{region}_price25ma']]
    else:
        graph_df=graph_df.join(region_df[f'{region}_price25ma'])

graph_df.tail()
```
