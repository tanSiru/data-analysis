```python
import pandas as pd

"""
unrelated but "%matplotlib tk" is a way to make the plot pop up for matplotlib
df is short for dataframe
dataframe: objects which are the rows and cols of our datas
series: a single column
panel: mutltiple dataframes
"""
df = pd.read_csv("avocado.csv")
```

```python
#this is used to see snippets of the data file
#the default will show the first 5 rows
#to change it just a enter a int
df.head()
```

```python
#same as df.head() but insteads show the end
df.tail()
```

```python
#same as df.head() but backwards, the amount shown can also be changed by enterin a int
df.tail()
```

```python
#we can access a specific column by doing it similar a dict
#.head() can also be used
#df.AveragePrice.head() but it's very uncommon and since how its got confusing to see 
df["AveragePrice"].head()
```

```python
#this is used to get a specific category within a column
#within the df dateframe, in the 'region' column get all the regions with the name of "Albany"
albany_df = df[df['region']=='Albany']
albany_df.head(100)
```

```python
#the index of a column is similar to a unique identifier in a database, its not always the case but it would be optimal
#it describes the relationships beetween all the columns
# albany_df.index(with this the index are kinda meaning less(it's just 1 2 3 4 ...))
#we use this to change the index and make it more meaningful
albany_df.set_index("Date")
```

```python
#this is because when just doing albany_df.set_index("Date") it doesn't commit the change 
#so we are reassigning albany_df to the change 
albany_df=albany_df.set_index("Date")
```

```python
albany_df
```

```python
#this is an alternative to change and commit the index change
albany_df.set_index("Date",inplace=True)
```

```python
#we can immediately plot the date
albany_df['AveragePrice'].plot()
```

```python
#on the previous plot pandas doesn't realize that our Date column is an actual column of dates
#this makes sure pandas know that our column "Date" consists of actual dates by convertiniig them to datetime objeects
df["Date"] = pd.to_datetime(df["Date"])
albany_df["AveragePrice"].plot()
```

```python
#Moving average: take the previous n point and take that average
#what were doing here is a moving average
albany_df['AveragePrice'].rolling(25).mean().plot()
```

```python
#when a chart appears like the previous plot it usually is because thee data is out of order
albany_df.index
#it might seem like it's in order but it starts in late december then descends, but then we later see 2018 appear 
```
